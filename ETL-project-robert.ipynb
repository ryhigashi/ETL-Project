```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import scipy.stats as st
```

```python
steve = pd.read_csv('API_NY.GDP.MKTP.KD.ZG_DS2_en_csv_v2_2445381.csv',low_memory=False, skiprows=4)
cory = pd.read_csv('API_NY.ADJ.NNTY.PC.CD_DS2_en_csv_v2_2459832.csv',low_memory=False, skiprows=4)
```

```python
median_df = pd.DataFrame(cory)
median_df.head()
```

```python
gdp_growth_df = pd.DataFrame(steve)
gdp_growth_df.head()
```

```python
years=[]
syear = 1960
for year in range(60):
    years.append(f"{syear + year}")
```

```python
countries = ["Australia", "Brazil", "Canada", "China", "France", "Germany", "India", "Italy", "Japan", "Mexico", "Russia", "Spain", "Switzerland", "United Kingdom", "United States"]
```

```python
df_gdp_list_start=[]
df_median_list_start=[]
df_gdp_list_trans = []
df_median_list_trans = []
for country in countries:
    df_gdp_list_start.append(gdp_growth_df.loc[gdp_growth_df['Country Name'] == country])
    df_median_list_start.append(median_df.loc[gdp_growth_df['Country Name'] == country])

for item in df_gdp_list_start:  
    df_gdp_list_trans.append(pd.melt(item, id_vars=['Country Name'], value_vars=['1960',
 '1961','1962','1963','1964','1965','1966','1967','1968','1969','1970','1971','1972','1973',
 '1974','1975','1976','1977','1978','1979','1980','1981','1982','1983','1984','1985','1986',
 '1987','1988','1989','1990','1991','1992','1993','1994','1995','1996','1997','1998','1999',
 '2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012',
 '2013','2014','2015','2016','2017','2018','2019']))

for item in df_median_list_start:  
    df_median_list_trans.append(pd.melt(item, id_vars=['Country Name'], value_vars=['1960',
 '1961','1962','1963','1964','1965','1966','1967','1968','1969','1970','1971','1972','1973',
 '1974','1975','1976','1977','1978','1979','1980','1981','1982','1983','1984','1985','1986',
 '1987','1988','1989','1990','1991','1992','1993','1994','1995','1996','1997','1998','1999',
 '2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012',
 '2013','2014','2015','2016','2017','2018','2019']))
    
```

```python
df_median_list_trans[0].head()
```

```python
countries_cleaned_median_df = pd.DataFrame()
countries_combined_median_df = pd.concat(df_median_list_trans)
countries_cleaned_median_df['country_name'] = countries_combined_median_df['Country Name']
countries_cleaned_median_df['year'] = countries_combined_median_df['variable'].astype(int)
countries_cleaned_median_df['median_income_adjusted'] = countries_combined_median_df['value'].fillna(0).astype(float).round(2)
countries_cleaned_median_df
```

```python
countries_cleaned_gdp_df = pd.DataFrame()
countries_combined_gdp_df = pd.concat(df_gdp_list_trans)
countries_cleaned_gdp_df['country_name'] = countries_combined_gdp_df['Country Name']
countries_cleaned_gdp_df['year'] = countries_combined_gdp_df['variable'].astype(int)
countries_cleaned_gdp_df['change_gdp'] = countries_combined_gdp_df['value'].fillna(0).astype(float).round(2)
countries_cleaned_gdp_df
```
